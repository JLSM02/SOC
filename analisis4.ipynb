```python
import numpy as np
import matplotlib.pyplot as plt
```

Gr√°fico de actividad sobre tiempo, ficheros FrameData

```python
limit = 12
nPoints = 150
min = 0.3
max = 1
completationLevels = np.linspace(min, max, nPoints)

amount = 32**2 * limit * completationLevels
sMean = np.array([])
for k in amount:
    data = np.loadtxt("Datos\sim004\FrameData_limit"+str(limit)+"_statistic"+str(int(k))+"_32x32.dat", delimiter=",", dtype=int)

    t = data[:, 0] #time
    s = data[:, 1] #size
    sDensity = s/32**2 #activation density
    if k in [amount[45], amount[67], amount[100]]:
    #if k in amount[65::20]:
        plt.plot(t, sDensity, label=rf"N / $L^2$ = {k/32**2: .2f}")

    sMean = np.append(sMean, np.mean(s[400,])) #mean size after transition

plt.title("Series temporales de tama√±os de avalanchas")
plt.xlabel("Tiempo (frames)")
plt.ylabel("Densidad de toplings (#toplings / #celdas)")
plt.xlim(0,120)
plt.ylim(0, 0.4)
plt.legend()
plt.show()
```

```python
limit = 12
nPoints = 150
min = 0.3
max = 1
completationLevels = np.linspace(min, max, nPoints)

amount = 32**2 * limit * completationLevels
sMean = np.array([])
for k in amount:
    data = np.loadtxt("Datos\sim004\FrameData_limit"+str(limit)+"_statistic"+str(int(k))+"_32x32.dat", delimiter=",", dtype=int)

    t = data[:, 0] #time
    s = data[:, 1] #size
    sDensity = s/32**2 #activation density
    if k in [amount[45], amount[67], amount[100]]:
    #if k in amount[65::20]:
        plt.plot(t, sDensity, label=rf"N / $L^2$ = {k/32**2: .2f}", marker='o')

    sMean = np.append(sMean, np.mean(s[400,])) #mean size after transition

plt.title("Series temporales de tama√±os de avalanchas")
plt.xlabel("Tiempo (frames)")
plt.ylabel("Densidad de toplings (#toplings / #celdas)")
plt.xscale("log")
plt.yscale("log")
plt.xlim(1,100)
plt.legend()
plt.show()
```

```python
plt.scatter(x=amount/32**2, y = sMean/32**2, s=10, label="L=32")

plt.title("Cambio de fase y punto cr√≠tico")
plt.xlabel("Densidad de arena (#arena / #celdas)")
plt.ylabel("<Densidad de toplings> (#toplings / #celdas)")
plt.legend()
plt.show()
```
